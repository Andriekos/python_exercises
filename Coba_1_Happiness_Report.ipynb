```python
import seaborn as sns
from pylab import savefig
```

```python
plt.figure(figsize=(25,20))
svm = sns.heatmap(df_17.corr(), annot=True, lw=1, cmap='ocean_r', fmt='.2f')

plt.suptitle('Map Correlation', fontsize=12)


figure = svm.get_figure()
figure.savefig('heatmap_17.png',dpi=600)
plt.show()
```

```python
df_17.info()
```

```python
plt.clf()

plt.figure(figsize=(16,15))

plt.subplot(3,1,1)
plt.scatter(df_17['Happiness.Score'],
            df_17['Economy..GDP.per.Capita.'], color='g')
plt.xlabel('Happiness Score')
plt.ylabel('Economy')

plt.subplot(3,1,2)
plt.scatter(df_17['Happiness.Score'],
            df_17['Family'], color='b')
plt.xlabel('Happiness Score')
plt.ylabel('Family')

plt.subplot(3,1,3)
plt.scatter(df_17['Happiness.Score'],
            df_17['Health..Life.Expectancy.'])
plt.xlabel('Happiness Score')
plt.ylabel('Health')

plt.suptitle('ECONOMY / FAMILY / HEALTH',fontsize=18)

plt.tight_layout()
plt.savefig('Corr.png',dpi=200)
plt.show()
```

```python
regions = df_16.Region
all_regions = set(regions)

print(all_regions)

print(len(all_regions))
```

```python
regions_color ={'Southeastern Asia':'red','Australia and New Zealand':'green',
                'Middle East and Northern Africa':'blue','Sub-Saharan Africa':'grey',
                'Southern Asia':'pink','North America':'yellow',
                'Western Europe':'orange','Central and Eastern Europe':'purple',
                'Eastern Asia':'aqua','Latin America and Caribbean':'olive'
}
```

```python
type(regions_color)
```

```python
colors = []
for i in df_16['Region']:
    colors.append(regions_color[i])
```

```python
plt.clf()

plt.figure(figsize=(20,10))

plt.scatter(df_16['Economy (GDP per Capita)'],
            df_16['Health (Life Expectancy)'],
            s=(df_16['Happiness Score']**4), alpha=0.5, c=colors
           )
plt.grid(True)

plt.xlabel('Economy')
plt.ylabel('Health')

plt.suptitle('Health Economy graph with size as Happiness score and colors as Region',
            fontsize=18)

plt.savefig('scatter_16.png',dpi=400)
plt.show()
```
